# Total-Debt-To-GDP Comparison

If China's GDP is only 39% of its official level, this means that official debt-to-gdp should be multiplied by a factor of 1/0.39.

```python
china_debt_to_gdp_official = 270 # https://www.google.com/search?q=china+total+debt+to+gdp&rlz=1C1VDKB_nlBE1024BE1024&oq=china+&aqs=chrome.0.69i59j69i57j69i60l3j69i65l3.1151j0j7&sourceid=chrome&ie=UTF-8
japan_debt_to_gdp_official = 164.38 +  221.07 # source https://www.imf.org/external/datamapper/datasets/GDD private + public debt
usa_debt_to_gdp_official = 180.83 + 119.01
china_gdp_estimated_size = 0.39
```

```python
china_debt_to_gdp_estimates = china_debt_to_gdp_official * (1/china_gdp_estimated_size)
```

```python
china_debt_to_gdp_estimates
```

Data for bar chart in video. 

```python
pd.DataFrame({'China Official': [china_debt_to_gdp_official], 'China Estimated': [china_debt_to_gdp_estimates],
              'Japan': [japan_debt_to_gdp_official], 'USA': [usa_debt_to_gdp_official],
             }).T.to_csv('china_debt.csv ')
```
