```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import world_bank_data as wb
import xml.etree.ElementTree as ET
import quandl
from IPython.display import Image
```

# Plotting Night Lights, China vs. India vs. South Korea

## GDP data from World Bank

```python
countries = wb.get_countries()
countries['name'][countries['name'] == 'China']
```

```python
GDP_LCU_China = pd.DataFrame(wb.get_series('NY.GDP.MKTP.KD.ZG', country='CHN', simplify_index=True))
GDP_LCU_China.index = GDP_LCU_China.index.astype(int)
GDP_LCU_India = pd.DataFrame(wb.get_series('NY.GDP.MKTP.KD.ZG', country='IND', simplify_index=True))
GDP_LCU_India.index = GDP_LCU_India.index.astype(int)
GDP_LCU_Korea = pd.DataFrame(wb.get_series('NY.GDP.MKTP.KD.ZG', country='KOR', simplify_index=True))
GDP_LCU_Korea.index = GDP_LCU_Korea.index.astype(int)
# select time frame that corresponds with night light dataset
GDP_LCU_China = GDP_LCU_China['NY.GDP.MKTP.KD.ZG'].loc[1993:2018]
GDP_LCU_India = GDP_LCU_India['NY.GDP.MKTP.KD.ZG'].loc[1993:2018]
GDP_LCU_Korea = GDP_LCU_Korea['NY.GDP.MKTP.KD.ZG'].loc[1993:2018]
```

### Plot GDP Growth

```python
fig, ax = plt.subplots(1, 1, figsize=(12,7))

ax.plot(GDP_LCU_China.index, GDP_LCU_China, label='China')
ax.plot(GDP_LCU_India.index, GDP_LCU_India, label='India')
ax.plot(GDP_LCU_Korea.index, GDP_LCU_Korea, label='Korea')

ax.set_ylabel('%', fontsize='16')
ax.set_xlabel('Year', fontsize='16')

ax.legend(loc='best', fontsize='16')
ax.set_title('World Bank GDP Growth Annual %', fontsize='28')
```

# Night Light data from Li et al. (2020)

I use the night-lights dataset from Li et al. 2020 that was provided in the supplementary materials from Martines (2022). 

https://www.nature.com/articles/s41597-020-0510-y

https://www.journals.uchicago.edu/doi/suppl/10.1086/720458

I first clean the data so that it only contains years and country name.

```python
nl = pd.read_csv('Harmo_zonalstat_1992_2018.csv')
years = ['Y{}_mean'.format(x) for x in range(1993, 2019)]
headers = ['CNTRY_NAME'] + years 
nl = nl[headers]
nl = nl.set_index('CNTRY_NAME')
nl.columns = [x for x in range(1993, 2019)]
nl = nl.T
nl.head()
```

To compare to GDP growth, I take the percentage change for each country.

```python
China_Night_Lights = nl['China'].pct_change()
India_Night_Lights = nl['India'].pct_change()
Korea_Night_Lights = nl['South Korea'].pct_change()
```

### Plot Night Light Growth

```python
fig, ax = plt.subplots(1, 1, figsize=(12,7))

ax.plot(China_Night_Lights.index, China_Night_Lights, label='China')
ax.plot(India_Night_Lights.index, India_Night_Lights, label='India')
ax.plot(Korea_Night_Lights.index, Korea_Night_Lights, label='Korea')

ax.set_ylabel('1993 = 100', fontsize='16')
ax.set_xlabel('Year', fontsize='16')

ax.legend(loc='best', fontsize='16')
ax.set_title('Li et al. (2020) Harmonized Nights Lights Index', fontsize='28')
```

# Plot them against each other

```python
fig, ax = plt.subplots(1, 1, figsize=(12,7))

ax.scatter(China_Night_Lights, GDP_LCU_China, label='China')
ax.scatter(India_Night_Lights, GDP_LCU_India, label='India')
#ax.scatter(Korea_Night_Lights, GDP_LCU_Korea, label='Korea')

ax.set_ylabel('GDP Growth', fontsize='16')
ax.set_xlabel('Night-Light Growth', fontsize='16')

ax.legend(loc='best', fontsize='16')
ax.set_title('China & India Night Light Growth vs. Reported GDP Growth', fontsize='28')
```

### Zoom in on cluster

```python
fig, ax = plt.subplots(1, 1, figsize=(12,7))

ax.scatter(China_Night_Lights, GDP_LCU_China, label='China')
ax.scatter(India_Night_Lights, GDP_LCU_India, label='India')
#ax.scatter(Korea_Night_Lights, GDP_LCU_Korea, label='Korea')

ax.set_ylabel('GDP Growth', fontsize='16')
ax.set_xlabel('Night-Light Growth', fontsize='16')

ax.legend(loc='best', fontsize='16')
ax.set_title('China & India Night Light Growth vs. Reported GDP Growth', fontsize='28')

ax.set_xlim(-0.2, 0.4)
ax.set_ylim(0, 15)
```
