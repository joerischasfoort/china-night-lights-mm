```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import binsreg
```

I take authoritarian Night Lights data that Martinez (2022) used to make his first figure and drop missing values. 

```python
data = pd.read_csv('Data_Figure1short.csv')
data = data.dropna(subset=['gdp14_growth', 'dn13_growth'])
```

Next, I split data into authoritarian and non-authoritarian 

```python
autocracies = data.loc[data['autocracyFH'] == 1.0]
democracies = data.loc[data['autocracyFH'] != 1.0]
```

```python
autocracies.head(2)
```

Then, because there are too many data points for a scatter plot. I use bins to summarize them (just like Martinez did). I followed this helpful tutorial https://towardsdatascience.com/goodbye-scatterplot-welcome-binned-scatterplot-a928f67413e4


```python
def binscatter(**kwargs):
    # Estimate binsreg
    est = binsreg.binsreg(**kwargs)
    
    # Retrieve estimates
    df_est = pd.concat([d.dots for d in est.data_plot])
    df_est = df_est.rename(columns={'x': kwargs.get("x"), 'fit': kwargs.get("y")})
    
    # Add confidence intervals
    if "ci" in kwargs:
        df_est = pd.merge(df_est, pd.concat([d.ci for d in est.data_plot]))
        df_est = df_est.drop(columns=['x'])
        df_est['ci'] = df_est['ci_r'] - df_est['ci_l']
    
    # Rename groups
    if "by" in kwargs:
        df_est['group'] = df_est['group'].astype(df[kwargs.get("by")].dtype)
        df_est = df_est.rename(columns={'group': kwargs.get("by")})

    return df_est
```

```python
# use binned scatterplot like in paper: 
auth_est = binscatter(x='dn13_growth', y='gdp14_growth', data=autocracies, ci=(3,3))
dem_est = binscatter(x='dn13_growth', y='gdp14_growth', data=democracies, ci=(3,3))
```

```python
auth_est
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12,7))

ax.scatter(dem_est['dn13_growth'], dem_est['gdp14_growth'], label='Democracies')
ax.scatter(auth_est['dn13_growth'], auth_est['gdp14_growth'], label='Autocracies')


ax.set_ylabel('GDP Growth %', fontsize='16')
ax.set_xlabel('Night-Light Growth %', fontsize='16')

ax.legend(loc='best', fontsize='16')
ax.set_title('Democracies & Autocracies Night Light Growth vs. Reported GDP Growth', fontsize='28')

ax.set_xlim(-20, 25)
ax.set_ylim(0, 8)
```
